```python
where_to_df = city_weather_data_df.set_index("City")
where_to_df
```

```python
hasaki_df = where_to_df.loc["Hasaki"]
hasaki_loc = hasaki_df[["Lat", "Lng"]]
hasakis = pd.DataFrame(hasaki_df)
hasaki = hasakis.transpose()
hasaki
```

```python
katsuura_df = where_to_df.loc["Katsuura"]
katsuura_loc = katsuura_df[["Lat", "Lng"]]
katsuuras = pd.DataFrame(katsuura_df)
katsuura = katsuuras.transpose()
katsuura
```

```python
shimoda_df = where_to_df.loc["Shimoda"]
shimoda_loc = shimoda_df[["Lat", "Lng"]]
shimodas = pd.DataFrame(shimoda_df)
shimoda = shimodas.transpose()
shimoda
```

```python
shingu_df = where_to_df.loc["Shingu"]
shingu_loc = shingu_df[["Lat", "Lng"]]
shingus = pd.DataFrame(shingu_df)
shingu = shingus.transpose()
shingu
```

# create dataframe with the 4 cities

```python
four_cities = [shingu, shimoda, katsuura, hasaki]

four_cities
```

```python
four_location_df = pd.concat(four_cities)
four_locations_df = pd.DataFrame(four_location_df)
four_locations_df
```

```python
# Declare a list that is to be converted into a column 
city = ['Shingu', 'Shimoda', 'Katsuura', 'Hasaki'] 
  
```

```python
# Using 'Address' as the column name 
# and equating it to the list 
four_locations_df['city_name'] = city
four_locations_df
```

```python
info_box_template = """
<dl>
<dt>Hotel Name</dt><dd>{Hotel Name}</dd>
<dt>City</dt><dd>{city_name}</dd>
<dt>Country</dt><dd>{Country}</dd>
<dt>Current Weather</dt><dd>{Current Description} & {Max Temp} °F</dd>
</dl>
"""
```

```python
# Store the DataFrame Row.
hotel_info = [info_box_template.format(**row) for index, row in four_locations_df.iterrows()]
```

```python
hotel_info
```

```python
locations = four_locations_df[["Lat", "Lng"]]

fig = gmaps.figure()
marker_layer = gmaps.marker_layer(locations, info_box_content=hotel_info)
travel_japan = gmaps.directions_layer(
        katsuura_loc, shimoda_loc, waypoints=[(hasaki_loc),(shingu_loc)],
        travel_mode='DRIVING', show_markers=False)
fig.add_layer(marker_layer)
fig.add_layer(travel_japan)


fig
```
